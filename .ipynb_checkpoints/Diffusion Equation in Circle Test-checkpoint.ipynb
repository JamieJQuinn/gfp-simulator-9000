````python
mesh = Mesh(50, min_size=0.005, min_angle=22)
fig, axis = plt.subplots(figsize=(10,10))
tr.plot(axis, **(mesh.triangulation))
plt.show()
````

````python
def calculate_area(p1, p2, p3):
    return 0.5*((p2[0]-p1[0])*(p3[1]-p1[1]) - (p2[1]-p1[1])*(p3[0]-p1[0]))

def is_von_neumman(interp):
    return interp < 0.5

def von_neumann_bc_value(interp):
    return 0
#     return 0.1*np.sin(2*np.pi*interp)

def calculate_von_neumann_bc(mesh):
    bc_vector = np.zeros(mesh.n_vertices)
    for vertex in mesh.boundary_vertices:
        if is_von_neumman(mesh.get_interp(vertex)):
            bc_vector[vertex] = 0.5*von_neumann_bc_value(mesh.get_interp(vertex))    
    return bc_vector

def calculate_body_force(vertex):
#     if vertex[0]**2 + vertex[1]**2 < 0.01:
#         return 1.0
#     else:
#         return 0.0
    return 4
#     return 1 - (vertex[0]**2 + vertex[1]**2)

def assemble_body_force(mesh):
    force_vector = np.zeros(mesh.n_vertices)
    for triangle in mesh.triangulation['triangles']:
        area = mesh.calculate_area(triangle)
        for index in triangle:
            force_vector[index] += 1/6.0 * area *calculate_body_force(mesh.get_pos(index))
    return force_vector

stiffness = mesh.assemble_stiffness()

stiffness[mesh.boundary_vertices,:] = 0.0
stiffness[:,mesh.boundary_vertices] = 0.0
for vertex in mesh.boundary_vertices:
    stiffness[vertex, vertex] = 1.0

bc_vector = calculate_von_neumann_bc(mesh)
force_vector = assemble_body_force(mesh)
    
force_vector[mesh.boundary_vertices] = 0.0

rhs_vector = bc_vector + force_vector

soln = np.linalg.solve(stiffness, rhs_vector)

fig, axis = plt.subplots(figsize=(10,10))
mesh.plot(soln, axis)
````

````python
error = [(soln[i] - (1.0 - vertex[0]**2 - vertex[1]**2))**2 for i, vertex in enumerate(mesh.triangulation['vertices'])]
fig, axis = plt.subplots(figsize=(10,10))
mesh.plot(error, axis)
````
